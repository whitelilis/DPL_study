```python
import gym
import gym_anytrading

env = gym.make("forex-v0", frame_bound=(50, 100), window_size = 10)
ob = env.reset()

while True:
    action = env.action_space.sample()
    ob, reward, done, info = env.step(action)
    print('-' * 50)
    print(info)
    print('-' * 50)
    if done:
        print("DONE---------")
        print(info)
        break

env.render_all()
```
